```python
import pandas as pd
```

```python
data = pd.read_csv("Polynomial.csv")
```

```python
data.head(4)
```

```python
x = data.iloc[:,:-1]
y = data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(criterion="squared_error",splitter="best",max_depth=20)
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
df = {"criterion":["squared_error","friedman_mse","absolute_error","poisson"],
     "splitter":["best","random"],
     "max_depth":[i for i in range(2,20)]}
```

```python
4*2*18
```

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=10)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```
