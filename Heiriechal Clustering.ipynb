```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("KNN Clustering.csv")
```

```python
data.head(4)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,19):
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,19)],wcss,marker="o")
plt.xlabel("no of cluster")
plt.ylabel("WCSS")
plt.grid(axis = "x")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn = KMeans(n_clusters=3)
data["Predict"] = kmn.fit_predict(data)
```

```python
kmn.labels_
```

```python
silhouette_score(data,labels=kmn.labels_)
```

```python
sns.pairplot(data=data,hue="Predict")
plt.show()
```

```python
ss = []
no_c= [j for j in range(2,19)]
for i in range(2,19):
    km1 = KMeans(n_clusters=i)
    km1.fit(data)
    ss.append(silhouette_score(data,km1.labels_))
```

```python
plt.plot(no_c,ss)
plt.xlabel("no of cluster")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```
